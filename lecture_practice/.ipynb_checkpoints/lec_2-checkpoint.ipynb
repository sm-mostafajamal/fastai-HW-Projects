# SGD

```python
import matplotlib.pyplot as plt
from fastai import *
import torch
import numpy as np
%matplotlib inline
```

```python
n = 100
```

```python
x = torch.ones(n,2)
x[:,0].uniform_(-1.,1)
x[:5]
```

```python
a = torch.tensor([4.,2]); a
```

```python
y = x@a + torch.rand(n); y[:5]
```

```python
plt.scatter(x[:,0],y)
```

```python
def mse(y_hat, y): return ((y_hat - y)**2).mean()
```

```python
a = torch.tensor([-1.,1])
```

```python
y_hat = x@a
mse(y_hat, y)
```

```python
plt.scatter(x[:,0], y_hat)
plt.scatter(x[:,0], y)

```

```python
import torch.nn as nn

a = nn.Parameter(a); a
lr = 1e-1
for t in range(100): update()
```

```python
def update():
    y_hat = x@a
    loss = mse(y, y_hat)
    if t % 10 == 0: print(loss)
    loss.backward()
    with torch.no_grad():
        a.sub_(lr * a.grad)
        a.grad.zero_()
        a.grad.zero_()
```

```python
plt.scatter(x[:,0],y)
plt.scatter(x[:,0],x@a.detach());
```

```python
from matplotlib import animation, rc
rc('animation', html='jshtml')
```

```python
a = nn.Parameter(torch.tensor([-1.,1]))

fig = plt.figure()
plt.scatter(x[:,0], y, c='orange')
line, = plt.plot(x[:,0], x@a.detach())
plt.close()

def animate(i):
    update()
    line.set_ydata(x@a.detach())
    return line,

animation.FuncAnimation(fig, animate, np.arange(0, 100), interval=20)

```
