```python
from fastai.vision import *
from fastai.metrics import error_rate
```

# Looking at Data

```python
path = Path('../input/intel-image-classification/seg_train/seg_train'); path.ls()
```

```python
np.random.seed(30)
data = ImageDataBunch.from_folder(path ,ds_tfms = get_transforms(), valid_pct=0.2, size = 224, bs = 64).normalize(imagenet_stats)
```

```python
data.classes
```

```python
data.show_batch(rows=3, figsize=(15,12))
```

# Training resnet34

```python
learn = create_cnn(data, models.resnet34, metrics = error_rate, model_dir="/tmp/model/")
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

# Result

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_top_losses(9, figsize=(15,12))
```

```python
interp.most_confused(min_val=2)
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi =60)
```
